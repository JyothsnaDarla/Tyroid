```python
import pandas as pd
df=pd.read_csv("/content/tyroid.zip")
df
```

```python
feature_cols=['age','T3','TT4','T4U','FTI']
features=df[feature_cols]
features
```

```python
Target=df.Classes
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(features,Target)
```

```python
c=input("enter age ")
d=input("enter T3 ")
e=input("enter TT4 ")
b=input("enter T4U ")
f=input("enter FIT ")
pred=model.predict([[c,d,e,b,f]])
if pred =='Negative':
  print("negative")
else:
  print("hypothyroid")
```

```python
from sklearn.tree import export_graphviz,plot_tree
```

```python
plot_tree(model)
```

```python
feature_cols=['age','T3','TT4','T4U','FTI']
features=df[feature_cols]
features
```

```python
Target=df.Classes
```

```python
from sklearn.ensemble import RandomForestClassifier
ml=RandomForestClassifier(n_estimators=100)
```

```python
ml.fit(features,Target)
```

```python
c=input("enter age ")
d=input("enter T3 ")
e=input("enter TT4 ")
b=input("enter T4U ")
f=input("enter FIT ")
pred=ml.predict([[c,d,e,b,f]])
if pred =='Negative':
  print("negative")
else:
  print("hypothyroid")
```

```python
import os
output_dir="tree_visualizations"
os.makedirs(output_dir,exist_ok=True)
```

```python
ml.estimators_
```

```python
for i,tree in enumerate(ml.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=["age","T3","TT4","T4U","FTI"],
                    class_names=[str(cls) for cls in ml.classes_], filled=True, rounded=True)
```

```python
command= f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree{i}visualization saved to {tree_png_file}")
```
